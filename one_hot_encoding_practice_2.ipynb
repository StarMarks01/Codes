```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import sklearn as slice
import seaborn as sns
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df.shape
```

```python
null_var = df.isnull().sum()/df.shape[0]*100
```

```python
dc = null_var[null_var>20].keys()
```

```python
df2 = df.drop(columns=dc)
```

```python
fill = df2.select_dtypes(include=(int,float)) 
```

```python
df = df2.fillna(fill.mean())
```

```python
dummy = pd.get_dummies(df)
dummy
```

```python
boolval = dummy.select_dtypes('bool')
boolval
```

```python
int_val = boolval.astype(int)
```

```python
df[dummy].replace(int_val)
```
