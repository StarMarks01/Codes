```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state = 42,test_size = 0.2)
```

```python
scaler.fit(xtrain)
xtrainscaled = scaler.transform(xtrain)
xtestscaled = scaler.transform(xtest)
```

```python
model1.fit(xtrainscaled,ytrain)
lins = model1.score(xtestscaled,ytest)
model1.predict(xtestscaled)
```

```python
poly.fit(xtrain)
xtrainpoly = poly.transform(xtrain)
xtestpoly = poly.transform(xtest)
model2.fit(xtrainpoly,ytrain)
wpolys = model2.score(xtestpoly,ytest)
model2.predict(xtestpoly)
```

```python
logic.fit(xtrainscaled,ytrain)
logics = logic.score(xtestscaled,ytest)
logic.predict(ypred)
```
