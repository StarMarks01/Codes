```python
df2 = df.copy()
df2 = df.drop(columns = ['Email','Address'])
def Outliers(columns, df):
    q1 = df2[columns].quantile(0.25)
    q3 = df2[columns].quantile(0.75)
    iqr = q3 - q1
    lowerlimit = q1 - (1.5 * iqr)
    upperlimit = q3 + (1.5 * iqr)
    # df2.loc[(df2[columns]>upperlimit) ,columns] = upperlimit
    # df2.loc[(df2[columns]<lowerlimit) ,columns] = lowerlimit
    return df[columns].clip(lowerlimit, upperlimit)
cols = ['Avg. Session Length', 'Time on App', 'Time on Website','Length of Membership', 'Yearly Amount Spent']
for i in cols:
    df2[i] = Outliers(i,df2)
```
